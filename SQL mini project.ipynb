```python
import os
import sqlite3
from sqlite3 import Error
```

```python
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        #print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn
```

```python
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)
```

```python
def main():
    database = "data/sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#Q10: Produce a list of facilities with a total revenue less than 1000.
    #The output of facility name and total revenue, sorted by revenue. Remember
    #that there's a different cost for guests and members! 
query1 = """
        with m as (
            SELECT f.facid, monthlymaintenance *
            (julianday(max(starttime)) - julianday(min(starttime)))/30 as totmaintenance
            from bookings b
            join facilities f 
            on b.facid = f.facid
            group by f.facid)
        select name,  
            round(sum(slots * (case when memid = 0 then guestcost
                          else membercost end)) - totmaintenance, 2) as totrevenue
        from bookings b
        join facilities f
            on b.facid = f.facid
        join m
            on m.facid = b.facid
        group by name
        having totrevenue < 1000
        order by totrevenue
        """
cur.execute(query1)
 
rows = cur.fetchall()
 
for row in rows:
    print(row)
```

```python
# Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order 
query1 = """
        SELECT m.firstname || ' ' || m.surname || ' recommended by ' || r.firstname || ' ' || r.surname
        FROM members m
        join members r
            on m.recommendedby = r.memid
        order by m.surname, m.firstname
        """
cur.execute(query1)
 
rows = cur.fetchall()
 
for row in rows:
    print(row)
```

```python
#Q12: Find the facilities with their usage by member, but not guests 
query1 = """
        SELECT name, firstname || ' ' || surname, sum(slots*.5)
        FROM facilities f
        join bookings b
            on f.facid = b.facid
        join members m
            on m.memid = b.memid
        where b.memid <> 0
        group by name, b.memid
        order by name, surname, firstname
        """
cur.execute(query1)
 
rows = cur.fetchall()
 
for row in rows:
    print(row)
```

```python
#Q13: Find the facilities usage by month, but not guests 

#Q12: Find the facilities with their usage by member, but not guests 
query1 = """
        SELECT name, strftime('%m',starttime) as month, firstname || ' ' || surname, sum(slots*.5)
        FROM facilities f
        join bookings b
            on f.facid = b.facid
        join members m
            on m.memid = b.memid
        where b.memid <> 0
        group by name, month, b.memid
        order by name, month, surname, firstname
        """
cur.execute(query1)
 
rows = cur.fetchall()
 
for row in rows:
    print(row)
```
